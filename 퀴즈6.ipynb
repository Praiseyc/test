```python
for pin in range(10000):
    trial = str(pin).zfill(4)  # 0000 형식으로 채움
    print("비밀번호 시도:", trial)
    if trial == "5732":  # 예: 실제 비밀번호
        print("비밀번호 찾음:", trial)
        break
```

### 1. 알고리즘 이름
- Brute Force (무차별 대입법)

### 2. 동작 방식
- 가능한 모든 경우(0000 ~ 9999)를 순차적으로 생성하여 도어락의 실제 비밀번호와 일일이 비교해 정답을 찾는 방식

### 3. 장점 / 단점
- 장점 : 구현이 매우 간단 / 정답이 존재하면 반드시 찾을 수 있음 (완전탐색 보장)
- 단점 : 매우 비효율적 (시간복잡도: O(N)) / 조합 수가 많아질수록 실행 시간이 폭발적으로 증가 / 실시간/보안 상황에서는 부적절하다.
